# 크롤링1 - 항목

```python
def cate_crawler(start_date, end_date):

    data_dict = []                                                                              # 수집한 데이터 딕셔너리 형태 저장 공간

    page_num = 1                                                                                # 페이지 카운트

    while True:
        craw1_url = "https://www.work24.go.kr/cm/openApi/call/hr/callOpenApiSvcInfo310L01.do"   # url 주소

        craw1_params = {                                                                        # 파라미터 설정
        "authKey": "e4c7fa60-8400-4d6c-8fd9-80552b74b17f",                                      # key 값
        "returnType": "JSON",                                                                   # 저장형태 (JSON / XML)
        "outType": 1,                                                                           # 출력형태 (1: 리스트 / 2: 상세)
        "pageNum": f"{page_num}",                                                               # 페이지 넘버
        "pageSize": 100,                                                                        # 페이지 출력건수
        "crseTracseSe": "C0104",                                                                # 훈련유형
        "srchTraStDt": start_date,                                                              # 시작일
        "srchTraEndDt": end_date                                                                # 마감일
        }

        response = requests.get(craw1_url, params = craw1_params)                                # requests 요청
        craw2_dump = response.json()                                                             # 요청받은 데이터 JSON 형태로 변경

        if not craw2_dump or not craw2_dump.get("srchList"):                                     # 수집한 데이터가 None값이면 while문 정지
            print(f"No more data found on page {page_num}. Stopping collection.")
            break

        data_dict.append(craw2_dump["srchList"])                                                 # 수집한 데이터 data_dict로 최종적으로 append

        page_num += 1                                                                            # 수집이 완료되면 페이지 넘버 증가

    return data_dict
```

```python
cate_craw_data = cate_crawler(20200101, 20301231)
```

```python
id_list = cate_df[["trprId", "trainstCstId", "trprDegr"]]
id_list
```

```python
craw2_url = "https://www.work24.go.kr/cm/openApi/call/hr/callOpenApiSvcInfo310L02.do"
common_params = {
    "authKey": "e4c7fa60-8400-4d6c-8fd9-80552b74b17f",
    "returnType": "JSON",
    "outType": 1,
}

result = []

for _, row in id_list[:10].iterrows():
    craw2_params = common_params.copy()
    craw2_params.update({
        "srchTrprId": row["trprId"],
        "srchTrprDegr" : row["trprDegr"],
        "srchTorgId" : row["trainstCstId"],
    })

    try:
        response = requests.get(craw2_url, params = craw2_params)
        response.raise_for_status()
        data = response.json()

        result.append(data)
    except:
        print("error")
```

```python
result_dump = []
for i in range(len(result)):
    merged_dict = {**result[i]["inst_base_info"], **result[i]["inst_detail_info"]}
    result_dump.append(merged_dict)
```
